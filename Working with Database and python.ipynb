```python
import logging
import pymongo
logging.basicConfig(filename="carbon_nanotubes.log", level="logging.debug", format='%(asctime)s %(levelname)s %(message)s')
try:
    logging.info("Establishing a connection to MongoDB")
    client = pymongo.MongoClient("mongodb+srv://mongodb:mongodb@cluster0.kmuuod1.mongodb.net/?retryWrites=true&w=majority")
    db = client.test
    logging.info(db)
except Exception as e:
    logging.exception(e)

logging.info("Creating Database connection and create collection")

try:
    db1 = client["CarbonNanoDB"]
    con = db1["CarbonNano"]
    logging.info("Datase and collection created succesfully")
except Exception as e:
    logging.error(e)

    
```

```python
import csv
list1 = []
try:
    logging.info("Openning a csv file carbon_nanotubes")
    with open("carbon_nanotubes.csv", "r") as file:
        data = csv.reader(file, delimiter='\n')
        for i in data:
            list1.append(i[0].split(';'))
except Exception as e:
    logging.error(e)
    
```

```python
try:
    logging.info("Writing csv data into new formated csv")
    with open("updated_carbon_nanoptubes.csv", "w", newline='') as data:
        writer = csv.writer(data)
        writer.writerows(list1)
except:
    logging.error("Issue in writing list data to csv file")
    logging.exception(e)
```
